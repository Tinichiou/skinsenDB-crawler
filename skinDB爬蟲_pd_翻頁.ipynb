```python
import xlwings as xw
wb = xw.Book('CCD-Batch-Search_2022-04-18_02_56_52.csv')
```

```python
sheet = wb.sheets[0]
```

```python
SMILES = sheet.range("E2:E24").value  #可用range.end.row 代替數值
```

```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
```

```python
loop = len(SMILES)//10
final = len(SMILES)%10
print(loop)
print(final)
```

```python
PATH=r"C:\Users\Tini\Downloads\chromedriver_win32\chromedriver.exe"  
```

```python
SMILESouts = []
ss_scores = []
ssresults = []
ADs = []
Acyls = []
MCs = []
SN2s = []
SNArs = []
Schiffs = []
```

```python
driver = webdriver.Chrome(PATH)
driver.get("https://cwtung.kmu.edu.tw/skinsensdb/predict")
search = driver.find_element_by_id("batchSmilesTextArea")
#輸入要查詢的SMILES
#在欄位中輸入SMILES enter鍵換行
for n in SMILES:
    search.send_keys(n)
    search.send_keys(Keys.RETURN)
time.sleep(5)

#在batch add按鈕左鍵單擊
actions = ActionChains(driver)
batchgo = driver.find_element_by_xpath('//*[@id="tab-content-1"]/div/md-content/div/div[1]/div[2]/md-content/div[1]/md-card/md-card-actions/button')
actions.click(batchgo).perform() 
time.sleep(7)

#pandas
df = pd.read_html(driver.page_source)[0]
print(df.head())
```
