```python
import xlwings as xw
wb = xw.Book('CCD-Batch-Search_2022-04-18_02_56_52.csv')
```

```python
sheet = wb.sheets[0]
```

```python
SMILES = sheet.range("E2:E24").value  #可用range.end.row 代替數值
```

```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
```

```python
loop = len(SMILES)//10
final = len(SMILES)%10
print(loop)
print(final)
```

```python
PATH=r"C:\Users\Tini\Downloads\chromedriver_win32\chromedriver.exe"  
```

```python
SMILESouts = []
ss_scores = []
ssresults = []
ADs = []
Acyls = []
MCs = []
SN2s = []
SNArs = []
Schiffs = []
import time
```

```python
for k in range(loop):
    driver = webdriver.Chrome(PATH)
    driver.get("https://cwtung.kmu.edu.tw/skinsensdb/predict")
    search = driver.find_element_by_id("batchSmilesTextArea")
    #輸入要查詢的SMILES LIST
    #在欄位中輸入SMILES enter鍵換行
    for i in range(0,10):
        search.send_keys(SMILES[(10*k)+i])
        search.send_keys(Keys.RETURN)
    time.sleep(5)
    #在batch add按鈕左鍵單擊
    actions = ActionChains(driver)
    batchgo = driver.find_element_by_xpath('//*[@id="tab-content-1"]/div/md-content/div/div[1]/div[2]/md-content/div[1]/md-card/md-card-actions/button')
    actions.click(batchgo).perform() 
    time.sleep(7)
    # 1. 輸出的SMILES 供檢查
    for i in range(1,11):
        outs = driver.find_element_by_xpath('//*[@id="tab-content-1"]/div/md-content/div/div[3]/div[2]/md-content/md-card/md-card-content/md-table-container/table/tbody/tr[%d]/td[1]'%(i))
        SMILESouts.append(outs.text)
    for n in SMILESouts:
        print(n)

    # 2. sensitizer score

    for i in range(1,11):
        score = driver.find_element_by_xpath('//*[@id="tab-content-1"]/div/md-content/div/div[3]/div[2]/md-content/md-card/md-card-content/md-table-container/table/tbody/tr[%d]/td[2]/span'%(i))
        ss_scores.append(score.text)
    for n in ss_scores:
        print(n)

    # 3. ssicon_result

    for i in range(1,11):
        icon = driver.find_element_by_xpath('//*[@id="tab-content-1"]/div/md-content/div/div[3]/div[2]/md-content/md-card/md-card-content/md-table-container/table/tbody/tr[%d]/td[2]/md-icon'%(i))
        ssresult = icon.get_attribute('ng-if').replace("res.Label==","")
        ssresults.append(ssresult)
    for n in ssresults:
        print(n)

    # 4. AD

    for i in range(1,11):
        AD = driver.find_element_by_xpath('//*[@id="tab-content-1"]/div/md-content/div/div[3]/div[2]/md-content/md-card/md-card-content/md-table-container/table/tbody/tr[%d]/td[3]'%(i))
        ADs.append(AD.text)
    for n in ADs:
        print(n)

    # 5. Acyl transfer agent

    for i in range(1,11):
        Acyl = driver.find_element_by_xpath('//*[@id="tab-content-1"]/div/md-content/div/div[3]/div[2]/md-content/md-card/md-card-content/md-table-container/table/tbody/tr[%d]/td[4]/md-icon'%(i))
        Acylresult = Acyl.get_attribute('ng-if').replace('res.AcylTransferAgent==','')
        Acyls.append(Acylresult)
    for n in Acyls:
        print(n)

    # 6. Michael acceptor

    for i in range(1,11):
        MC = driver.find_element_by_xpath('//*[@id="tab-content-1"]/div/md-content/div/div[3]/div[2]/md-content/md-card/md-card-content/md-table-container/table/tbody/tr[%d]/td[5]/md-icon'%(i))
        MCresult = MC.get_attribute('ng-if').replace('res.MichaelAcceptor==','')
        MCs.append(MCresult)
    for n in MCs:
        print(n)

    # 7. SN2

    for i in range(1,11):
        SN2 = driver.find_element_by_xpath('//*[@id="tab-content-1"]/div/md-content/div/div[3]/div[2]/md-content/md-card/md-card-content/md-table-container/table/tbody/tr[%d]/td[6]/md-icon'%(i))
        SN2result = SN2.get_attribute('ng-if').replace('res.SN2==','')
        SN2s.append(SN2result)
    for n in SN2s:
        print(n)

    # 8. SNAr

    for i in range(1,11):
        SNAr = driver.find_element_by_xpath('//*[@id="tab-content-1"]/div/md-content/div/div[3]/div[2]/md-content/md-card/md-card-content/md-table-container/table/tbody/tr[%d]/td[7]/md-icon'%(i))
        SNArresult = SNAr.get_attribute('ng-if').replace('res.SNAr==','')
        SNArs.append(SNArresult)
    for n in SNArs:
        print(n)

    # 9.Schiff base

    for i in range(1,11):
        Schiff = driver.find_element_by_xpath('//*[@id="tab-content-1"]/div/md-content/div/div[3]/div[2]/md-content/md-card/md-card-content/md-table-container/table/tbody/tr[%d]/td[8]/md-icon'%(i))
        Schiffresult = Schiff.get_attribute('ng-if').replace('res.SchiffBaseFormation==','')
        Schiffs.append(Schiffresult)
    for n in Schiffs:
        print(n)
    
    driver.quit()
```
